## Plot Types 
```
Distribution plots display a single continous feature and help visualize properties such as deviatgion and average values . 

3 main distribution plot types :
    - Rug Plot
    - Histogram 
    - KDE PLot 

```    

**Rug Plot**
```
A Rug Plot is the simplest distribution plot and mereky adds a dash or tick line for every single value . 
The y-axis does not really have a meaning. 
```

**KDE**

KDE - stands for Kernel Density Estimation.
```
It is a method of estimating a probablity density function of a random variable . 
In simpler terma - it is a way of estimating a continous probablity curve for a finite data sample .  
```
##### Note: We can change the kernal and bandwidth used which can make your KDE show more or less of the variance contained in the data .

````python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns  
````

````python
df = pd.read_csv('/Applications/My_Laborarory/Experiment/myenv/source/Seaborn/data/dm_office_sales.csv')
df.head()
````

````python
sns.rugplot(x='salary',data=df,height=0.5)
````

````python
plt.figure(figsize=(5,8),dpi=200)
sns.rugplot(x='salary',data=df,height=0.5)
````

````python
# DO NOT USE DISTPLOT -- DEPRECIATED 
#sns.displot
#sns.histplot
## kwargs stands for key words arguments 
sns.set(style='ticks')
sns.displot(data=df,x='salary',bins=20,color='red',edgecolor='blue')
````

````python
sns.histplot(data=df,x='salary',color='red',edgecolor='black')
````

````python
sns.set(style='ticks')
sns.displot(data=df,x='salary',bins=20,color='red',edgecolor='blue',kde=True,rug=True)
````

````python
# just the KDE plot
sns.kdeplot(data=df,x='salary')
````

````python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
````

````python
sample_ages
````

````python
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
sample_ages
````

````python
sns.rugplot(data=sample_ages,x='age')
````

````python
sns.displot(data=sample_ages,x='age',rug=True,bins=20,edgecolor='black',kde=True)
````

````python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100],bw_adjust=0.2,shade=True)
````
