# Categories Plots 

The catagorical plots discussed here will display a statistical metrics per a catagory 

For example - mean value per category or a count of the number of rows per catagory .

```It is the visualization equivalent of a groupby() call```

##### The 2 main types of plots for this are - 
```
      1. countplot() - Counts number of rows per category 
  
      2. barplot() - General form of displaying any chosen metric per category  
```

**1.countplot**

###### Counts number of rows per category
###### we can also do hue separation for each category 

**2. barplot**
   
###### General form of displaying any chosen metric per category  
###### The barplot is the general form that allows you to choose any measure or estimator for the y axis . 
######  We could plot the mean value and standard deviation per category instead . 

## Important Note ! 
```
    -- Be very careful with these plots , since the bar is filled and continous , a viewer may interpret continuity along the y-axis which may be incorrect ! 
    -- Always make sure to add additional labeling and explanation for these plots ! 
```

# LETS DO SOME CODING 

````python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
````

````python
df = pd.read_csv('/Applications/My_Laborarory/Experiment/myenv/source/Seaborn/data/dm_office_sales.csv')
df.head()
````

````python
df['division'].value_counts()
````

````python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='division')
````

````python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='division')
#plt.ylim(90,260)
````

````python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='level of education',hue='division')
````

````python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='level of education',hue='division',palette='Set2',edgecolor='blue')
````

````python
df['level of education'].value_counts()
````

````python
plt.figure(figsize=(10,4))
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,errorbar='sd',hue='division')

plt.legend(bbox_to_anchor=(1.05,1))
````
