# Comparison PLots 

Comparison plots are essentially 2-dimensional version of the plots we've learned about so far . 

The two main types discussed here : 
    -- jointplot()
    -- pairplot()

#####  JOINTPLOT 

We can map histograms to each feature of a scatterplot to clarify the distributions within eacvh feature . 
We can also adjust the scatterplot to bea hex plot or a 2D KDE plot . 

##### PAIRPLOT 

The pairplot() is a quick way to compare all numerical columns in a DataFrame . 
It automatically creates a histogram for each column and a scatterplot comparison between all possible combinations of columns . 

**Warning !** 
```
    -- Pairplot() can be CPU and RAM intensive for large DataFrames with many columns . 
    -- It is a good idea to first filter down to only columns you are interested in .  
```


````python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt 
import seaborn as sns 
````

````python
df = pd.read_csv('/Applications/My_Laborarory/Experiment/myenv/source/Seaborn/data/StudentsPerformance.csv')
df.head()
````

````python
sns.jointplot(data=df,x='math score',y='reading score',kind='hex')
````

````python
sns.jointplot(data=df,x='math score',y='reading score',kind='scatter',alpha=0.2)
````

````python
sns.jointplot(data=df,x='math score',y='reading score',kind='hist',alpha=0.2)
````

````python
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',fill=True)
````

````python
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',fill=True,hue='gender')
````

````python
sns.pairplot(data=df,hue='gender')
````

````python
sns.pairplot(data=df,hue='gender',diag_kind='hist')
````

````python
sns.pairplot(data=df,hue='gender',corner=True)
````
