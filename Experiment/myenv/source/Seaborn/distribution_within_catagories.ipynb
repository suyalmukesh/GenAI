# Categorical Plots 
```
We've explored distribution plots for a single feature , but what if we want to compare distributions across categories ? 

For example , instead of the distribution of everyone's salary , we can compare the distributions of salaries per level of education .

```

Distribution within Catagories 

```
    1. Boxplot 
    2. Violinplot 
    3. Swarmplot
    4. Boxenplot (Letter-Value Plot)
```    

**Boxplot** 
```
The Boxplot displays the distribution of a continous variable. 
It does this through the use of quartiles . 
Quartiles seperate out the data into 4 equal number of data points :
    25% of data points are in bottom quartile . 
    50th percentile (Q2) is the median . 

Boxplot quickly gives statistical distribution information in a visual format :     
Boxplot can be oriented vertically or horizontally 
```

**Violin PLot** 
```
The violin plot plays a similar role as the box plot .
It displays the probability density across the data using a KDE.
We can imagine it as a mirrored KDE plot . 
```

**Swarmplot** 
```
The swarmplot is very simple and simply shows all the data points in the distribution . 
For very large data sets , it won't show all the points , but will display the general distribution of them . 
```
**Boxenplot ( Letter-value plot )** 
```
It is a relatively new plot developed in 2011 by Heike Hofmann , Karen Kafadar , and Hadley Wickham . 
Its mainly designed as an expansion upon the normal box plot . 
```


````python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns  
````

````python
df = pd.read_csv('/Applications/My_Laborarory/Experiment/myenv/source/Seaborn/data/StudentsPerformance.csv')
df.head()
````

````python
plt.figure(figsize=(6,2),dpi=200)
sns.boxplot(data=df,y='math score',x='test preparation course',hue='test preparation course')
````

````python
plt.figure(figsize=(10,4))
sns.boxplot(data=df,y='reading score',x='parental level of education',hue='parental level of education')
````

````python
plt.figure(figsize=(10,4))
sns.boxplot(data=df,y='reading score',x='parental level of education',hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05,0.5))
````

````python
plt.figure(figsize=(10,8))
sns.boxplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set2')

plt.legend(bbox_to_anchor=(1.05,0.5))
````

````python
plt.figure(figsize=(10,8))
sns. violinplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set2')

plt.legend(bbox_to_anchor=(1.5,0.5))

````

````python
plt.figure(figsize=(10,8))
sns. violinplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set2',split=True)

plt.legend(bbox_to_anchor=(1.5,0.5))
````

````python
plt.figure(figsize=(10,8))
sns. violinplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set2',split=True,inner=None)

plt.legend(bbox_to_anchor=(1.5,0.5))
````

````python
plt.figure(figsize=(10,8))
sns. violinplot(data=df,x='reading score',y='parental level of education',hue='test preparation course',palette='Set2',split=True,inner='quartile')

plt.legend(bbox_to_anchor=(1.5,0.5))
````

````python
plt.figure(figsize=(10,8))
sns. violinplot(data=df,x='reading score',y='parental level of education',split=True,inner='stick')
````

````python
plt.figure(figsize=(10,8))
sns. violinplot(data=df,x='reading score',y='parental level of education',bw_method=2)
````

````python
sns.swarmplot(data=df,x='math score')
````

````python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df,x='math score',y='gender',size=2,hue='test preparation course',dodge=True)

plt.legend(bbox_to_anchor=(1.35,0.5))
````

````python
sns.boxenplot(x='math score',y='test preparation course',data=df,hue='gender')
````
