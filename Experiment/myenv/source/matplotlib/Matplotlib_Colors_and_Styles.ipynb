```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
x
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='blue')

# RGB HEX Code 
# Google search Hex Color Picker 

ax.plot(x,x,color='#e0a910',label='x v x ')

ax.plot(x,x+1,color='purple',label='x vs x+1')


ax.legend()
```

```python
## Line Width (linewidth or lw )
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='blue')

# RGB HEX Code 
# Google search Hex Color Picker 

ax.plot(x,x,color='#e0a910',lw=10)

ax.plot(x,x+1,color='purple',linewidth=8)


ax.legend()
```

```python
# Line Styles 

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e0a910',lw=10,ls='--')

ax.plot(x,x+1,color='purple',linewidth=8,linestyle='-.')


ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x,color='#e0a910',lw=5,)


lines[0].set_dashes([5,2,10,2])   # 5 points 2 spaces 10 points 2 spaces etc 

lines[0].set_dashes([1,2,1,2,10,2])
```

```python
# Markers 
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e0a910',lw=3,marker='o',ls='--',ms=10)

ax.plot(x,x+1,color='red',lw=3,marker='+',ls='--',markersize=20)


ax.plot(x,x+5,color='blue',lw=3,marker='s',ls='--',markersize=20,markerfacecolor='red',markeredgewidth=8,markeredgecolor='orange')
# How to decide markers ?

## Google search matplotlib.markers  
```
