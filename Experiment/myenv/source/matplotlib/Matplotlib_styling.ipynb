````python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
````

````python
x = np.linspace(0,11,10)
````

````python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X^2') 

#ax.legend(loc='lower left')
#ax.legend(loc=0)
ax.legend(loc=(1.1,0.5))
````
