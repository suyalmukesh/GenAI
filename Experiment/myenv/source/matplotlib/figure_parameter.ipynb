# Matplotlib - Figure Parameter

#### figsize 
#### dpi 
#### bbox_inches

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)  # To get linearly space objects 
b = a**4
```

```python
fig = plt.figure(figsize=(2,2),dpi=100)  # To set the high resolution 

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

fig.savefig('output/new_fig.png',bbox_inches='tight')
```
