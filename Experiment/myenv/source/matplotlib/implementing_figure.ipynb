```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)  # To get linearly space objects 
b = a**4

```

```python
x = np.arange(0,10)
y = 2*x
```

```python
# These all command should into the same shell 

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y);

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)
```

```python
fig = plt.figure()
#Large axis
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#Small axis 
axes2 = fig.add_axes([1,1,2,2])
axes2.plot(x,y)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in Power of 4')


axes.plot(x,y)
```
