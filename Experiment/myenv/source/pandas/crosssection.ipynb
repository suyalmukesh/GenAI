```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/mpg.csv")

df.head()
```

```python
numeric_df = df.select_dtypes(include=['number'])

mean_values = numeric_df.groupby(df['model_year']).mean()

mean_values
```

```python
year_cyl = numeric_df.groupby(['model_year','cylinders']).mean()
year_cyl
```

```python
year_cyl.index.names
```

```python
year_cyl.index.levels
```

# xs - cross-section

```python
year_cyl.xs(key=70,level='model_year')

# xs only takes one key 
```

```python
four_cyl = year_cyl.xs(key=4,level='cylinders')
four_cyl
```

```python
# if we need multiple keys , we use loc 

year_cyl.loc[[70,80]]
```
