# Working with single argument function (single column)

```python
def last_four(num):
    return int(str(num)[-4:])
```

```python
last_four(1234567)
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
df
```

```python
df['total_bill']
```

```python
# Create a custom function to return 1 dollar sign or 2 or 3 

def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
df
```

# Working with multiple arguments function (multiple columns)

```python
df['total_bill'].apply(lambda num: num * 2)
```

```python
df.head()
```

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'    
```

```python
quality(16.99,1.01)
```

```python
df['Quality'] = df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
df
```

# Vectorize  (np.vectorize)

```python
## Now how to make it faster 

df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
df
```
