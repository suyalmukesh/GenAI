# Group By 
```
A groupby() operation allows us to examine data on a per category basis .
```

````python
import numpy as np
import pandas as pd
````

````python
dtype = {'mpg':'float64',
         'cylinders':'int64',
         'displacement':'float64',
         'horsepower':'string',
         'weight':'int64',
         'acceleration':'float64',
         'model_year':'int64',
         'origin':'int64',
         'name':'string' 
         }

df = pd.read_csv("data/mpg.csv",dtype=dtype)

df.head()
````

````python
df['horsepower'].fillna(0)
````

````python
df['horsepower'] = df['horsepower'].replace({'?':'0'})
````

````python
df['horsepower'].unique()
````

````python
df.astype({'horsepower':'int64'}).dtypes
````

````python
df.dtypes
````

````python
df['model_year'].unique()
````

````python
df['model_year'].value_counts()
````

````python
print(df['origin'].unique())
print(df['origin'].dtype)
````

````python
numeric_df = df.select_dtypes(include=['number'])

mean_values = numeric_df.groupby(df['model_year']).mean()

mean_values
````

````python
numeric_df.groupby('model_year').mean()
````

````python
numeric_df.groupby('model_year').mean()['origin']
````

````python
df['horsepower'] = df.horsepower.astype(int)
````
